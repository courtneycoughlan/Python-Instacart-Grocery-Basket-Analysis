3. Aggregating data

```python
# Create a subset
df = df_ords_prods_merge[:1000000]
```

```python
# Aggregate mean of order_number grouped by department_id for entire dataframe
df_ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
# Group data
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# View data with new column
df.head(10)
```

```python
# Aggregate data using loyalty flag
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Display frequency counts
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# View data by certain columns
df[['user_id', 'loyalty_flag', 'prices', 'add_to_cart_order']].head(60)
```

```python
# Group data
df['mean_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View data with new column
df.head(10)
```

```python
# Create a spending flag
df.loc[df['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df.loc[df['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# View frequency counts
df['spending_flag'].value_counts(dropna = False)
```

```python
# View data with certain columns
df[['user_id', 'spending_flag', 'prices']].head(60)
```

```python
# Group data
df['median_days'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View data with new column
df.head(10)
```

```python
# Aggregate data with frequency flag
df.loc[df['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['median_days'] <= 20) & (df['median_days'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df.loc[df['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# View frequency counts
df['frequency_flag'].value_counts(dropna = False)
```

```python
# View data with certain columns
df[['user_id', 'frequency_flag', 'order_id']].head(60)
```

4. Export file as pkl

```python
# Export df at pkl file
df.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```
